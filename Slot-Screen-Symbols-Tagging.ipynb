# Въведение

Ротативките (известни също като слот машини) се една от най-популярните форми в съвременния хазарт. Първоначално те са представлявали механични устройства, които дават за награда някакъв десерт. Изработвани са от въртящи се барабани, които играчът завърта с дърпане на ръчка (от там и другото им название - едноръки бандити). Щом барабаните спрат своето въртене, пред играча се формира линии. В линиите може да има печалба, а може и да няма.

Съвременните ротативки са компютъризирани устройства или уеб базирани приложения, които имитират въртенето на механичните барабани с помощта на масиви от числа, изпълняващи своеобразната роля на виртуални барабани. Виртуалната среда дава много повече възможности от това, което позволява изработката на механични устройства. Поради тази причина, съвременните ротативки имат значително по-сложна механика и игрови сценарии. В общия случай, ротативките подреждат пред играча символи в три реда и пет колони. Веднъж подредени, символите формират печеливши комбинации и според таблицата на печалбите дават окончателна награда на играча.

Решаването на този проблем често е свързан с желанието на определени производители да наподобят игровата механика на чужди успешни заглавия.

# Използвани методи

Методът cv2.matchTemplate от библиотеката OpenCV представлява класически подход за съпоставяне на изображения, известен като template matching. Този метод се използва за откриване на даден шаблон (малко изображение) в по-голямо изображение чрез систематично сравнение на пикселни стойности. Основната идея е шаблонът да бъде „плъзган“ върху цялото изображение и за всяка позиция да се изчислява числова оценка, която показва степента на съвпадение между шаблона и съответния регион от изображението.

В конкретния случай ще се използва режимът cv2.TM_CCOEFF_NORMED, който реализира нормализирана крос-корелация с изваждане на средната стойност. Това е усъвършенстван вариант на корелационните методи, при който първо се премахва средната стойност на пикселите както от шаблона, така и от текущия прозорец в изображението. Тази стъпка има за цел да елиминира влиянието на общото осветление и контраст, като по този начин алгоритъмът става по-устойчив към промени в яркостта. След това се изчислява корелацията между двата набора от стойности, а резултатът се нормализира чрез стандартното отклонение. В резултат се получава стойност в интервала от -1 до +1, където +1 означава пълно съвпадение, 0 липса на зависимост, а -1 – противоположност.

Алгоритмично методът комбинира няколко ключови компонента. На първо място стои самата крос-корелация, която измерва линейната зависимост между два сигнала – в случая между шаблона и част от изображението. На второ място се прилага центриране чрез изваждане на средната стойност, което премахва постоянните компоненти в сигнала. След това се извършва нормализация, която осигурява съпоставимост на резултатите независимо от размера на интензитетите. Целият процес се реализира чрез механизъм на плъзгащ прозорец, при който шаблонът се премества пиксел по пиксел върху изображението. Това води до сравнително висока изчислителна сложност, особено при по-големи изображения или голям брой шаблони.

Този метод ще се използва за класификация на изображения чрез директно сравнение с набор от предварително подготвени шаблони. Всяко входно изображение се сравнява с всички шаблони, като за всеки от тях се изчислява карта на съвпадение. Чрез функцията cv2.minMaxLoc се извлича максималната стойност от тази карта, която служи като мярка за най-доброто съвпадение. Накрая се избира шаблонът с най-висока стойност и той се приема като клас на изображението.

Този подход е ефективен при строго контролирани условия, например когато всички изображения са с еднакъв размер и без значителни геометрични трансформации. Въпреки това той има ограничения, тъй като е чувствителен към мащабиране, ротация и деформации, както и към шум. Освен това методът не извлича абстрактни характеристики от изображенията, а работи директно с пикселни стойности, което го отличава от съвременните методи, базирани на машинно обучение.

## Практическа реализация

Поради ограниченият обем на тази курсова работа, само част от поставените задачи ще бъдат изпълнени. За целите на изследването са събрани група изображения, които са предварително записани в работна под директория. Автоматичното заснемане на игровите екрани е задача с определена сложност, която би била подходяща за разработка с по-голям обхват и повече ресурси за изпълнения.

Подреждането на етикетираните отрязъци е комбинаторна задача, която значително надхвърля сложността, характерна за една курсова работа. Такова реконструиране изисква сложен статистически анализ и комбинаторни оптимизационни алгоритми.

В настоящата курсова работа акцентът пада върху изолирането на отделните изображения от игрови символи и тяхното класифициране, като им се съпоставят подходящи етикети.

```python
import os
import cv2
import pandas as pd
from PIL import Image
from pathlib import Path
from collections import defaultdict
```

```python
os.getcwd()
data_folder = Path("./data")
template_folder = Path("./templates")
extraction_folder = Path("./extracted")
extraction_folder.mkdir(exist_ok=True)
```

## Зареждане на референтни еталони

Класификацията значително може да се улесни, ако бъдат заредени предварителни еталони, които да служат като референтни изображения.

```python
symbol_width = 95
symbol_height = 95

templates = {}
for template in template_folder.glob("*.png"):
    name = template.stem
    image = cv2.imread(str(template), cv2.IMREAD_COLOR_RGB)
    resized = cv2.resize(image, (symbol_width, symbol_height))
    templates[name] = resized
```

## Изрязване на символите

Без да се търси голяма прецизност, която в последствие може да се наложи за подобряване на класификацията, общото изображение на екрана се обхожда, определят се зоните за изрязване и се записват отделни файлове за всеки игрови символ. За всеки екран се оформя отделна под директория.

```python
reel_area = (243, 114) 
cols, rows = 5, 3

vertical_spacing = 30
horizontal_spacing = 6

def extract_symbols(image_path, output_root):
    image = Image.open(image_path)
    image_name = Path(image_path).stem 
    
    target_dir = Path(output_root) / image_name
    target_dir.mkdir(parents=True, exist_ok=True)
    
    for r in range(rows):
        for c in range(cols):
            left = reel_area[0] + (c * (symbol_width + vertical_spacing))
            top = reel_area[1] + (r * (symbol_height + horizontal_spacing))

            symbol = image.crop((left, top, left + symbol_width, top + symbol_height))
            symbol.save(target_dir / f"symbol_r{r}_c{c}.png")
```

## Прочитане на изображенията

Вече подготвените изображения на снети екрани се изчитат в последователен ред с помотща на билиотеката PIL. Ако се използва извикването на функцията image.show(), изображението бива визуализирано в програма за показване на изображения, аоциирана с този вид файлове в конкретната операционна система.

```python
for file_path in data_folder.glob("*.png"):
    extract_symbols(file_path, extraction_folder)
```

## Формиране на класове

След като отделните игрови символи са изолирани в под директории и собствени графични файлове, следва определянето на класовете в които могат да попаднат. Тъй като броят символи не е голям, то референтни еталони, за всеки символ, могат да се оформят в помощтна директория tamplates. Тази стъпка е извършена ръчно, но в бъдещи разработки може да се автоматизира, като се анализират множеството семпли.

Резултатът от класификацията се записва в същата директория, като етикетът е в текстов файл, съвпадащ с името на графичния файл, но с разширение за текст.

```python
def classify_images(folder_path):
    crop_folder = Path(folder_path)
    
    for crop_file in crop_folder.glob("*.png"):
        crop_image = cv2.imread(str(crop_file), cv2.IMREAD_COLOR_RGB)
        best_score = -1
        best_match = "Unknown"

        for name, template_image in templates.items():
            result = cv2.matchTemplate(crop_image, template_image, cv2.TM_CCOEFF_NORMED)
            _, max_value, _, _ = cv2.minMaxLoc(result)

            if max_value > best_score:
                best_score = max_value
                best_match = name

        tag = best_match #if best_score > 0.8 else "Uncertain"
        
        text_file = crop_file.with_suffix('.txt')
        
        with open(text_file, 'w') as file:
            file.write(tag)
```

## Обхождане на изолираните изображения

Всички директории с извадени изображения на символите трябва да бъдат обходени и класификацията трябва да бъде стартирана по отделно.

```python
for subfolder in extraction_folder.iterdir():
    if subfolder.is_dir():
        classify_images(subfolder)
```

## Честотен анализ за срещането на символите по клетки

След класификацията на отделните символи по клетките от съществено значение е честотата с кято всеки символ се среща във всяка от клетките. По естествен начин, символите даващи по-големи печалби трябва да влизат по-рядко в печеливши комбинации и съответно по-рядко се появяват в игралното поле.

```python
histogram = defaultdict(lambda: defaultdict(int))

for subdir in os.listdir(extraction_folder):
    subdir_path = os.path.join(extraction_folder, subdir)
    
    if os.path.isdir(subdir_path):
        for filename in os.listdir(subdir_path):
            if filename.endswith('.txt'):
                file_path = os.path.join(subdir_path, filename)
                
                row_name = os.path.splitext(filename)[0]
                
                with open(file_path, 'r') as f:
                    word = f.read().strip()
                    if word:
                        histogram[row_name][word] += 1

df = pd.DataFrame.from_dict(histogram, orient='index').fillna(0).astype(int)
df = df.sort_index(axis=0).sort_index(axis=1)

#print(df)
df.to_csv(extraction_folder /'symbol_frequencies.csv')
```

# Заключение

Резултатите от етикетирането са приложени в директория, създадена от скрипта за класификация. Направена е ръчно оценка за фалшиво положителни или фалшиво отрицателни резултати и в подбраното множество от заснети екрени не се установяват грешни класификации, дори когато символите са частично зачертани, защото участват в печеливша комбинация. 

Основни недостатъци на избрания метод за класифициране е, че е чувствителен към преоразмеряване на изображенията, ротация в изображенията и наличието на други видове деформации. При голямо количество референтни еталони, алгоритмите в този метод започват да изискват твърде много изчислително време. Тъй като става въпрос за числено сравнение на информацията от пикселите, не се влага никакво изследване за обектите от които са изградени изображенията.

При нужда от по-сложна класификация, удачни биха били методи с извличане на характеристики, като SIFT или ORB, конволюционни изкуствени невронни мрежи и сравняване на хистограми.
